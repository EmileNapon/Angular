Exercice 7


Crée une grille 2D (meshgrid) de x et y dans l’intervalle [-2, 2] avec 9 points, puis crée un
tableau Z = X² + Y².

```python
x=np.linspace(-2,2,9)
y=np.linspace(-2,2,9)

X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2
```
